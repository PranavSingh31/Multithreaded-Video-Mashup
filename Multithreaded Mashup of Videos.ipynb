```python
import os
import time
import threading
import glob
import moviepy.editor as mp
import time
```

```python
time = []
```

```python
def cut_and_merge(video_file_path, output_file_path):
    video = mp.VideoFileClip(video_file_path)
    clip = video.subclip(0, 20)
    clip.write_videofile(output_file_path, audio=False)
```

```python
def cut_and_merge_multi_thread(video_file_paths, output_file_path, num_threads=1):
    def process_chunk(chunk_index):
        start_index = chunk_index * chunk_size
        end_index = (chunk_index + 1) * chunk_size if chunk_index != num_chunks - 1 else len(video_file_paths)
        for i in range(start_index, end_index):
            video_file_path = video_file_paths[i]
            output_file_path = f"{i}.mp4"
            cut_and_merge(video_file_path, output_file_path)

    num_videos = len(video_file_paths)
    chunk_size = num_videos // num_threads
    num_chunks = num_videos // chunk_size + (num_videos % chunk_size != 0)
    threads = []

    for i in range(num_chunks):
        thread = threading.Thread(target=process_chunk, args=(i,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    # Merge all clips
    video_clips = [mp.VideoFileClip(f"{i}.mp4") for i in range(num_videos)]
    final_clip = mp.concatenate_videoclips(video_clips)
    final_clip.write_videofile(output_file_path, audio=False)
```

```python
if __name__ == "__main__":
    # Input files
    video_file_paths = ["../Downloads/Videos/0.mp4", "../Downloads/Videos/1.mp4", "../Downloads/Videos/2.mp4", "../Downloads/Videos/3.mp4", "../Downloads/Videos/0.mp4"]
    output_file_path = "output.mp4"
```

```python
start_time = time.time()
cut_and_merge_multi_thread(video_file_paths, output_file_path, num_threads=4)
end_time = time.time()
print(f"Four-threaded runtime: {end_time - start_time:.3f} seconds")
time.append(end_time-start_time)
```
